## Loading the Sample Dataset

We will work with a sample dataset of global economic indicators. Note: this is a **synthetic teaching dataset** for practice only.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the sample data
url = "https://raw.githubusercontent.com/GSIS-DS-Fall-2026/data-science-ai-global-decision-making/main/data/sample/global_indicators_sample.csv"
df = pd.read_csv(url)
df.head()
```

## Previewing the Data

Let's examine the structure and basic statistics of our dataset.

```python
# Basic info about the dataset
print(f"Number of observations: {len(df)}")
print(f"Countries: {df['country'].unique()}")
print(f"Years: {sorted(df['year'].unique())}")
```

## Creating a Simple Chart

Let's visualize GDP growth rates across countries in 2022.

```python
# Filter for 2022 and create a bar chart
df_2022 = df[df['year'] == 2022]
plt.figure(figsize=(10, 6))
plt.bar(df_2022['country'], df_2022['gdp_growth'])
plt.xticks(rotation=45, ha='right')
plt.xlabel('Country')
plt.ylabel('GDP Growth (%)')
plt.title('GDP Growth in 2022')
plt.tight_layout()
plt.show()
```

## Four Levels of Analysis

For each chart or finding, distinguish:

1. **Observation**: What you see in the data (the bars, numbers, or values)
2. **Pattern**: What the observation suggests (trend, relationship, or difference)
3. **Interpretation**: Why the pattern might exist (economic, political, or social factors)
4. **Decision claim**: What action the pattern suggests (policy, business, or research decisions)
