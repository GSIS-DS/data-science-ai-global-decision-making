```python
import pandas as pd
import matplotlib.pyplot as plt

path = '../../data/sample/global_indicators_sample.csv'
df = pd.read_csv(path)
df.head()
```

```python
df[['country','year','gdp_growth','region']].head(10)
```

```python
avg_growth = df.groupby('region', as_index=False)['gdp_growth'].mean().sort_values('gdp_growth', ascending=False)
ax = avg_growth.plot(kind='bar', x='region', y='gdp_growth', legend=False, figsize=(8,4), color='steelblue')
ax.set_title('Average GDP Growth by Region (Synthetic Data)')
ax.set_xlabel('Region')
ax.set_ylabel('GDP Growth (%)')
plt.tight_layout()
plt.show()
```
