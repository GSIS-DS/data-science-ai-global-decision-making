# Week 2: Working with Data in Python

In this notebook, you will learn essential pandas skills for working with global data.

## Loading Data

Start by loading our sample dataset.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/GSIS-DS-Fall-2026/data-science-ai-global-decision-making/main/data/sample/global_indicators_sample.csv"
df = pd.read_csv(url)
df
```

## Inspecting Columns

Use `.columns` to see all column names and `.dtype` to check data types.

```python
# See all column names
df.columns.tolist()
```

```python
# Check data types for each column
df.dtypes
```

## Selecting Data

Use brackets to select a single column, returning a Series.

```python
# Select one column
gdp_growth = df['gdp_growth']
gdp_growth
```

Use double brackets to select multiple columns, returning a DataFrame.

```python
# Select multiple columns
df[['country', 'year', 'gdp_growth']]
```

## Filtering Data

Use conditions to filter rows.

```python
# Filter for one country
korea = df[df['country'] == 'South Korea']
korea
```

```python
# Filter for GDP growth above 5%
high_growth = df[df['gdp_growth'] > 5]
high_growth
```

## Summary Statistics

Use `.describe()` for quick statistics on numeric columns.

```python
df.describe()
```

```python
# Mean for specific column
df['inflation'].mean()
```
