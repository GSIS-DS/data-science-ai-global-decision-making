### Global question
How did inflation and unemployment vary across regions in 2022?

```python
import pandas as pd

file_path = '../../data/sample/global_indicators_sample.csv'
year_of_interest = 2022

df = pd.read_csv(file_path)
df.head()
```

```python
df.columns.tolist()
```

```python
subset_2022 = df[df['year'] == year_of_interest].copy()
subset_2022[['country','region','inflation','unemployment_percent']].head()
```

```python
summary = subset_2022[['inflation','unemployment_percent']].describe().round(2)
summary
```
