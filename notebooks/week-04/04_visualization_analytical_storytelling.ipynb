## Good Visualization Principles

Every chart should include:
1. Clear title
2. Labeled axes with units
3. Source attribution
4. Accessible design (colorblind-friendly colors)
5. Honest scale (no misleading truncation)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/GSIS-DS-Fall-2026/data-science-ai-global-decision-making/main/data/sample/global_indicators_sample.csv"
df = pd.read_csv(url)
```

## Bar Chart Example

CO2 emissions per capita by region.

```python
plt.figure(figsize=(10, 6))
regional_co2 = df.groupby('region')['co2_per_capita'].mean().sort_values()
regional_co2.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('CO2 per capita (metric tons)')
plt.title('Average CO2 Emissions by Region (2020-2022)')
plt.tight_layout()
plt.show()
```

## Line Chart Example

GDP growth trends over time.

```python
plt.figure(figsize=(10, 6))
for country in ['South Korea', 'Japan', 'Germany', 'United States']:
    country_data = df[df['country'] == country].sort_values('year')
    plt.plot(country_data['year'], country_data['gdp_growth'], marker='o', label=country)
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')
plt.title('GDP Growth Trends (2020-2022)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Misleading Chart Example

This chart truncates the y-axis to exaggerate differences.

```python
# Misleading: truncated y-axis
plt.figure(figsize=(8, 5))
countries = ['South Korea', 'Japan', 'Germany']
values = [1.7, 1.9, 1.7]  # 2022 GDP growth
plt.bar(countries, values)
plt.ylim(1.5, 2.0)  # Misleading truncation
plt.ylabel('GDP Growth (%)')
plt.title('Notice how the truncated axis exaggerates differences') 
plt.show()
```
