# Misleading Visualization Example

This notebook shows common visualization mistakes.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/GSIS-DS-Fall-2026/data-science-ai-global-decision-making/main/data/sample/global_indicators_sample.csv"
df = pd.read_csv(url)
```

## Problem: Truncated Y-axis

```python
# Misleading: truncated axis makes small differences look large
plt.figure(figsize=(8, 5))
countries = ['South Korea', 'Japan', 'Germany']
values = [1.7, 1.9, 1.7]
plt.bar(countries, values)
plt.ylim(1.5, 2.0)  # Misleading!
plt.ylabel('GDP Growth (%)')
plt.title('Small Differences Exaggerated by Truncated Axis')
plt.show()
```
