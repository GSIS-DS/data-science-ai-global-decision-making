# Good Visualization Example

This notebook demonstrates proper chart design.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/GSIS-DS/data-science-ai-global-decision-making/main/data/sample/global_indicators_sample.csv"
df = pd.read_csv(url)
```

```python
plt.figure(figsize=(10, 6))
regional_gdp = df.groupby('region')['gdp_growth'].mean()
regional_gdp.plot(kind='bar', color='steelblue')
plt.xlabel('Region')
plt.ylabel('Average GDP Growth (%)')
plt.title('Average GDP Growth by Region (2020-2022)')
plt.axhline(y=0, color='gray', linestyle='-', linewidth=0.5)
plt.tight_layout()
plt.show()
```
